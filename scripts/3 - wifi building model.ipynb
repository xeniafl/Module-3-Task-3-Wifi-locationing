```python
import numpy as np
import pandas as pd
import scipy
from math import sqrt
import matplotlib.pyplot as plt
import mglearn
from plotnine import *

#estimators
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
wifidf = pd.read_csv('wifidfnew.csv')
```

```python
validationdf = pd.read_csv('validationdfnew.csv')
```

```python
#features
X_train = wifidf.iloc[:, 1:521]
X_train.head()
```

```python
y_train = wifidf.iloc[:, 524]
y_train.head()
```

```python
#Testing Set (X Testing)
X_validation = validationdf.iloc[:, 1:521]
X_validation.head()
```

```python
list(X_validation.columns)
```

```python
#Ground Truth (y_test) 
y_validation = validationdf.iloc[:, 524]
y_validation_count = len(y_validation.index)
print('The number of observations in the Y training set are:',str(y_validation_count))
y_validation.head()
```

```python
X_train.shape, X_validation.shape
```

```python
#Models
modelknn = KNeighborsClassifier(n_neighbors=2)
modellm = LogisticRegression()
modelsvm = LinearSVC()
modelrf = RandomForestClassifier()
```

```python
modelknn.fit(X_train, y_train)
```

```python
modellm.fit(X_train,y_train)
```

```python
modelsvm.fit(X_train,y_train)
```

```python
modelrf.fit(X_train,y_train)
```

```python
modelknn.score(X_train,y_train)
```

```python
modellm.score(X_train,y_train)
```

```python
modelsvm.score(X_train,y_train)
```

```python
modelrf.score(X_train,y_train)
```

```python
predictions = modelrf.predict(X_validation)
```
