```python
# We'll start with building prediction, as it's the easiest one:
```

```python
# Feature selection:
X_train = wifidf.iloc[:, 1:521]
y_train = wifidf.iloc[:, 524]
```

```python
#Testing Set (X Testing)
X_validation = validationdf.iloc[:, 1:521]
#y_validation = validationdf.iloc[:, 524]
```

```python
modellm = LogisticRegression()
modellm.fit(X_train,y_train)
```

```python
predictions = modellm.predict(X_validation)
predictions
```

```python
y_validation
```

```python
confusion = confusion_matrix(y_validation, predictions)
print("Confusion matrix:\n{}".format(confusion))
```

```python
accuracy = accuracy_score(y_validation, predictions)
accuracy
```

```python
predictions = pd.DataFrame(predictions)
predictions.columns = ['BUILDINGID']
wifidfwithbuilding = pd.concat([X_validation, predictions], axis=1)
wifidfwithbuilding.to_csv('validationwithbuilding.csv')
```

```python
wifidf = pd.read_csv('wifidfnew.csv')
validationdf = pd.read_csv('validationwithbuilding.csv')
#validationdf2 = pd.read_csv('validationdfnew.csv')
```

```python
X_train = wifidf.iloc[:, 1:521]
X_train = X_train.replace(100, -1000)
X_train = X_train.astype(float)
X_train = 10**(X_train/10)
y_train = wifidf.iloc[:, 523]
```

```python
validationdf
```

```python
X_validation = validationdf.iloc[:, 1:521]
X_validation = X_validation.replace(100, -1000)
X_validation = X_validation.astype(float)
X_validation = 10**(X_validation/10)
#y_validation = validationdf2.iloc[:, 523]
```

```python
modelknn = KNeighborsClassifier(n_neighbors = 4, weights = 'distance', p = 1)
modelknn.fit(X_train, y_train)
```

```python
predictions = modelknn.predict(X_validation)
predictions
```

```python
y_validation
```

```python
confusion = confusion_matrix(y_validation, predictions)
print("Confusion matrix:\n{}".format(confusion))
```

```python
accuracy = accuracy_score(y_validation, predictions)
accuracy
```

```python
predictions = pd.DataFrame(predictions)
predictions.columns = ['FLOOR']
submissionfile = pd.concat([wifidfwithbuilding, predictions], axis=1)
submissionfile.to_csv('submissionfile.csv')
```

```python
wifidf = pd.read_csv('wifidfnew.csv')
validationdf = pd.read_csv('validationwithbuilding.csv')
#validationdf2 = pd.read_csv('validationdfnew.csv')
```

```python
X_train = wifidf.iloc[:, 1:521]
X_train = X_train.replace(100, 0)
X_train[X_train < -78] = 1
X_train[X_train < 0] = 2
X_train = pd.concat([X_train, wifidf.iloc[:, 524]], axis=1)
y_train = wifidf.iloc[:, 522]
```

```python
X_validation = validationdf.iloc[:, 1:521]
X_validation = X_validation.replace(100, 0)
X_validation[X_validation < -78] = 1
X_validation[X_validation < 0] = 2
X_validation = pd.concat([X_validation, validationdf.iloc[:, 521]], axis=1)
#y_validation = validationdf2.iloc[:, 522]
```

```python
modelknn = neighbors.KNeighborsRegressor(n_neighbors = 7, weights = 'distance')
modelknn.fit(X_train, y_train)
```

```python
predictions = modelknn.predict(X_validation)
predictions
```

```python
y_validation
```

```python
MAE = sum(abs(y_validation - predictions))/len(predictions)
predRsquared = r2_score(y_validation,predictions)
```

```python
#Make Predictions
print('R Squared: %.3f' % predRsquared)
print('MAE: %.3f' % MAE)
```

```python
predictions = pd.DataFrame(predictions)
predictions.columns = ['LATITUDE']
submissionfile = pd.concat([submissionfile, predictions], axis=1)
submissionfile.to_csv('submissionfile.csv')
```
